```python
import pandas as pd

file_name = "flightsinUSA_2020.xlsx"
Flight = pd.ExcelFile(file_name)

Flight.sheet_names
```

```python
# select sheet page by sheet name
fact = Flight.parse('fact')
Dim_time = Flight.parse('Dim_time')
Dim_name = Flight.parse('Dim_name')
Dim_origin= Flight.parse('Dim_origin')
Dim_destination = Flight.parse('Dim_destination')
```

```python
table_merge = fact.merge(Dim_time, how ='left', left_on='time_ID',right_on='time_ID')
print(table_merge.shape)
table_merge.head()
```

```python
table_merge1 = table_merge.merge(Dim_name, how ='left', left_on='name_ID',right_on='name_ID')
print(table_merge1.shape)
table_merge1.head()
```

```python
table_merge2 = table_merge1.merge(Dim_origin, how ='left', left_on='origin_ID',right_on='origin_ID')
print(table_merge2.shape)
table_merge2.head()
```

```python
table_merge3 = table_merge2.merge(Dim_destination, how ='left', left_on='destination_ID',right_on='destination_ID')
print(table_merge3.shape)
table_merge3.head()
```

```python
table_merge3.to_excel('Flight_USA.xlsx',index=False)
```
